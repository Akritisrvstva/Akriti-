We will explore a customer churn dataset using Pandas, Matplotlib, and Seaborn libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

The first step is to read the dataset into a pandas dataframe.



```python
df= pd.read_csv('/content/Churn_Modelling (1).csv')
```

The below mwntioned code will display the boxplot graph using seaborn library. The graph is being plotted between geography and tenure. 

```python
sns.boxplot(data=df, x="Geography", y="Tenure")
plt.grid(True)
plt.show()
```

Let's take two columns Estimated Salary and Balance from our dataset and see what we can infer by plotting to scatter plot between estimated salary and balance. 

```python
plt.scatter(df.EstimatedSalary, df.Balance)
plt.show()
```

Let's take two columns Age and Balance from our dataset and see what we can infer by plotting to scatter plot between Age and balance. 

```python
df.plot.scatter(x="Age", y="Balance")
plt.show()
```

**Pair-Plot**-

Now, lets plot Pair Plots for the three columns we used in plotting Scatter plots. Well use the seaborn library for plotting Pair Plots.

```python
sns.pairplot(data = df, vars=['EstimatedSalary','Balance','Age'])
plt.show()
```

Below code will let us know about the datatype of every column in the dataset.

```python
df.dtypes
```

```python
df[['Geography','Gender','Exited']].groupby(['Geography','Gender']).agg(['mean','count'])
```

```python
fig , axs = plt.subplots(ncols=2, figsize=(12,6))
fig.suptitle("Distribution of Balance and Estimated Salary", fontsize=15)
sns.distplot(df.Balance, hist=False, ax=axs[0])
sns.distplot(df.EstimatedSalary, hist=False, ax=axs[1])
```

```python
df['Gender'].replace({'Male':0, 'Female':1}, inplace=True)
corr = df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap='Blues_r', annot=True)
```

```python
plt.figure(figsize=(6,6))
plt.title("Boxplot of the Age Column", fontsize=15)
sns.boxplot(y=df['Age'])
```
