FILL ALL THE NULL VALUES

```python
for i in ['director','cast','country','date_added','rating']:
    df[i] = df[i].fillna(df[i].mode()[0])
```

PLOT THE BOXPLOT GRAPH FOR THE SKEWNESS FOR THE ANALYSIS OF RELEASE YEAR

```python
plt.figure(figsize=(12,6))
print("Skewness is :",df.release_year.skew())
sns.boxplot(df.release_year)
df.release_year.describe()
plt.title(f"Analysis of Release year - Box Plot" , fontsize=20,
          color="orange")
```

**Observation**

The above graph is showing us the box plot of the release year of different tv shows and movies and the skewness is -3.6191659982330684

PLOT THE GRAPH FOR SHOWING THE SKEWNESS AND ANALYSIS OF RELEASE YEAR

```python
print("Skewness is :", df.release_year.skew())
plt.figure(figsize=(12,6))
sns.distplot(df.release_year)
df.release_year.describe()
plt.title(f"Analysis of release_year - skewness" , fontsize=20,
          color="orange")
```

**Observation**

The above graph is showing the skewness for the graph of release year , and the above graph is negatively skewed. And the skewness is -3.6191659982330684
